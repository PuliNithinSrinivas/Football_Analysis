```python
from ultralytics import YOLO # type: ignore
import numpy as np # type: ignore
from scipy.stats import entropy  # type: ignore # For statistical calculations
from sklearn.cluster import DBSCAN  # type: ignore # DBSCAN Clustering

# Load trained YOLOv8 model
model = YOLO('../models/best.pt')

# Run validation
metrics = model.val(data="c:/Users/megha/OneDrive/Desktop/football_analysis/traning/football-players-detection-1/data.yaml")

# Extract evaluation metrics
mAP50 = metrics.box.map50  # Mean Average Precision at IoU 0.5
precision = metrics.box.mp  # Precision (TP / (TP + FP))
recall = metrics.box.mr  # Recall (TP / (TP + FN))

# Calculate estimated accuracy using F1-score formula
accuracy = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

# Store all metrics in an array for Z-score calculation
metric_values = np.array([mAP50, precision, recall, accuracy])

# Calculate mean and standard deviation
mean_val = np.mean(metric_values)
std_dev = np.std(metric_values)

# Compute Z-Scores
z_scores = (metric_values - mean_val) / std_dev if std_dev > 0 else np.zeros_like(metric_values)

# Apply DBSCAN clustering on metric values
metric_values_reshaped = metric_values.reshape(-1, 1)  # Reshape for DBSCAN
dbscan = DBSCAN(eps=0.1, min_samples=1).fit(metric_values_reshaped)  # Fit DBSCAN model
clusters = dbscan.labels_  # Extract cluster labels

# Print results
print(f"mAP50: {mAP50}")  
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Estimated Accuracy: {accuracy}")
print(f"Z-Scores (mAP50, Precision, Recall, Accuracy): {z_scores}")
print(f"DBSCAN Clusters (mAP50, Precision, Recall, Accuracy): {clusters}")
```
